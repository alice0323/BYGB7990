```python
import pandas as pd
%matplotlib inline
```

```python
tweet=pd.read_csv('badbot.csv')
tweet.head()
```

```python
A = tweet['created_at']
```

```python
len(A)
```

```python
date = []
time = []
for i in range(0,139912):
    new_date = A[i].split(' ')[0]
    new_time = A[i].split(' ')[1]
    date.append(new_date)
    time.append(new_time)
```

```python
date
```

```python
pf = pd.DataFrame({'created_at':A,
                   'date':date,
                   'time':time})
```

```python
pf
```

```python
data = pd.merge(tweet, pf,on = 'created_at', how = 'inner') 
data
```

```python
D1 = data[['id','tweet_id','date','time','text']]
D1
```

```python
data.groupby('id').date.value_counts()
```

```python
tweet_id_count = data.groupby('id').tweet_id.value_counts()
K=pd.DataFrame(tweet_id_count)
```
